```python
import pandas as pd
import numpy as np
from pandas import read_csv
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
temperature_data = read_csv('city_temperature.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
```

```python
temperature_data
```

```python
temperature_data = temperature_data[temperature_data['Day'] > 0]
```

```python
temperature_data = temperature_data.reset_index()
```

```python
temperature_data = temperature_data.drop(['Region'], axis = 1)
```

```python
temperature_data['AvgTemperature'] = temperature_data['AvgTemperature'].map(lambda x: (x - 32) * 5/9)
```

```python
temperature_data['Date'] = np.nan
```

```python
temperature_data_RU_RO = temperature_data[(temperature_data['Country'] == 'Romania') | (temperature_data['Country'] == 'Russia')]
```

```python
def transform_date_to_one_var(df):
    date_columns_df = df[['Year', 'Month', 'Day']].copy()
    date = pd.to_datetime(date_columns_df)
    df['Date'] = date
    df = df.drop(['State', 'Month', 'Day', 'Year'], axis = 1)
    return df
```

```python
temperature_data_RU_RO = transform_date_to_one_var(temperature_data_RU_RO)
```

```python
temperature_data_RU_RO = temperature_data_RU_RO[(temperature_data_RU_RO['City'] == 'Moscow') | (temperature_data_RU_RO['City'] == 'Bucharest')]
```

```python
temperature_data_RU_RO = temperature_data_RU_RO.reset_index()
```

```python
temperature_data_RU_RO
```

```python
temperature_data_RU_RO = temperature_data_RU_RO[(temperature_data_RU_RO['AvgTemperature'] > -30)]
```

```python
temperature_data_RU_RO_2019 = temperature_data_RU_RO[temperature_data_RU_RO['Date'] > '2019']
```

```python
temperature_data_RU_RO_2019 = temperature_data_RU_RO_2019.reset_index()
```

```python
temperature_data_RU_RO_2019 = temperature_data_RU_RO_2019.drop('level_0', axis = 1)
```

```python
temperature_data_RO_2019 = temperature_data_RU_RO_2019[temperature_data_RU_RO_2019['Country'] == 'Romania']
```

```python
temperature_data_RO_2019.plot.line(x='Date', y='AvgTemperature', figsize = (20, 10))
```

```python
temperature_data_RO_2019_positive = temperature_data_RO_2019[temperature_data_RO_2019['AvgTemperature'] >= 0]
```

```python
temperature_data_RO_2019_positive.plot.area(x='Date', y='AvgTemperature', figsize = (20, 10))
```

```python
max_temp_df = temperature_data_RU_RO_2019[["Country", "AvgTemperature"]].groupby("Country", as_index=False).apply(lambda df:df.sort_values("AvgTemperature", ascending=False).head(1)).droplevel(0).sort_values("Country", ascending=False)
```

```python
max_temp_df.plot.bar(x='Country', y='AvgTemperature', rot=0)
```
